```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("data/flight_delays.csv")
```

```python
df.head()
```

```python
x = df["Month"]
y = df["NK"]
```

```python
plt.bar(x,y)
plt.show
```

```python
width = 0.5

plt.style.use('tableau-colorblind10')

plt.bar(x,y,color="m",width=0.5)

plt.title("Average Arrival Delay for Spirit Airlines Flights, by Months")
plt.xlabel("Months")
plt.ylabel("Attival Delay(in minutes)")

plt.show()
```

I notice that the highest delay on average per month is in june, and that the lowest is in the months september-november; with october being the lowest.

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("data/Taxi_Trip.csv")
df.head()
```

```python
x = df["trip_duration"].head(20)
y = df["tip"].head(20)
```

```python
plt.plot(x, y)
plt.show()
```

```python
trip_duration = df["trip_duration"].head(20)
tip = df["tip"].head(20)
total_fare = df["total_fare"].head(20)
plt.plot(trip_duration, tip)
plt.plot(trip_duration, total_fare)
```

```python
trip_duration = df["trip_duration"].head(20)
tip = df["tip"].head(20)
total_fare = df["total_fare"].head(20)
plt.plot(trip_duration, tip)
plt.plot(trip_duration, total_fare)
```

```python
trip_duration = df["trip_duration"].head(20)
tip = df["tip"].head(20)
total_fare = df["total_fare"].head(20)

plt.plot(trip_duration, tip,color="r", linestyle="--",linewidth=6,marker="o", label= "tip")
plt.plot(trip_duration, total_fare, color="g",linestyle=':',marker=".",linewidth=6, label= "total_fare" )

plt.title("Total Fare vs Tips Recieved by Trip Duration")
plt.xlabel("Trip Duration")
plt.ylabel("Amount payed(in USD)")

plt.legend()
plt.grid(True)
plt.show()
```

One thing I noticed about the dataset is that the fare increases at a higher rate than the tip. Although driving a longer distance will make a higher tip, the tip will not be proportional to what it wouldve been, had the trip been shorter.
